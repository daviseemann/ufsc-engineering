```python
import matplotlib.pyplot as plt
import numpy as np
```

Um determinado sistema responde a uma excita√ß√£o em degrau de
acordo com a seguinte equa√ß√£o:

$$ùëì(ùë°) = ùê¥ ‚ãÖ (1 ‚àí ùëí^{‚àíùë°/2})$$

Escreva um c√≥digo que plote tal fun√ß√£o e indique os tempos de resposta a 80%
e 99% do sistema (ou seja, indique quando a resposta atinge 80% e 99% do seu
valor de regime permanente).
Para plotar a fun√ß√£o e indicar os tempos de resposta a 80% e 99%, podemos usar o seguinte c√≥digo:


```python
t = np.linspace(0, 15, 1000)
A = 5


def f(t):
    return A * (1 - np.exp(-t / 2))


# Calculate response times
t_80 = t[np.where(f(t) >= 0.8 * A)[0][0]]
t_99 = t[np.where(f(t) >= 0.99 * A)[0][0]]

fig, ax = plt.subplots()
ax.plot(t, f(t), label="f(t)")
ax.annotate(
    f"t_80 = {t_80:.2f}",
    xy=(t_80, 0.8 * A),
    xytext=(t_80 + 1, 0.8 * A - 1),
    arrowprops=dict(arrowstyle="->", facecolor="red"),
    color="red",
)
ax.annotate(
    f"t_99 = {t_99:.2f}",
    xy=(t_99 - 1, 0.99 * A),
    xytext=(t_99 + 1, 0.99 * A - 1),
    arrowprops=dict(arrowstyle="->", facecolor="green"),
    color="green",
)
ax.set(xlabel="t", ylabel="f(t)", title="$F(t) = A(1 - e^{-t/2})$")
ax.legend()
ax.grid()
plt.show()
```
