```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Generate a sample signal:
np.random.seed(0)
t = np.linspace(0, 1, 500, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t) + np.random.normal(0, 0.5, t.shape)
# Design a Butterworth low-pass filter:
order = 4
cutoff = 6  # Cutoff frequency in Hz
nyquist = 0.5 * 500  # Nyquist frequency
```

```python
normal_cutoff = cutoff / nyquist
b, a = butter(order, normal_cutoff, btype="low", analog=False)
# Apply the filter to the signal
filtered_signal = filtfilt(b, a, signal)
# Plotting:
fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(t, signal, label="Original Signal", color="red", alpha=0.5)
ax.plot(t, filtered_signal, label="Filtered Signal", color="blue")
ax.set_title("Signal Filtering using SciPy")
ax.set_xlabel("Time [s]")
ax.set_ylabel("Amplitude")
ax.legend()
ax.grid(True)
plt.show()
```
